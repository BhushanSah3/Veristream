```python
import numpy as np
import pandas as pd
import tensorflow as tf
from transformers import TFAutoModel, AutoTokenizer  # TensorFlow-compatible Transformers
import matplotlib.pyplot as plt
from sklearn.metrics import ConfusionMatrixDisplay
from sklearn.model_selection import train_test_split
from sklearn.metrics import classification_report

```

```python
# Load Dataset
true_data = pd.read_csv('a1_True.csv')
fake_data = pd.read_csv('a2_Fake.csv')

# Generate labels True/Fake under new Target Column in 'true_data' and 'fake_data'
true_data['Target'] = ['True']*len(true_data)
fake_data['Target'] = ['Fake']*len(fake_data)

# Define a dictionary for label mapping
label_map = {"Fake": 1, "True": 0}

# Apply the label mapping to the 'Target' column in both DataFrames
true_data['label'] = true_data['Target'].replace(label_map)
fake_data['label'] = fake_data['Target'].replace(label_map)

# Merge 'true_data' and 'fake_data', by random mixing into a single df called 'data'
data = pd.concat([true_data, fake_data]).sample(frac=1).reset_index(drop=True)


# See how the data looks like
print(data.shape)
data.head()
```

```python
# Checking if our data is well balanced
label_size = [data['label'].sum(),len(data['label'])-data['label'].sum()]
plt.pie(label_size,explode=[0.1,0.1],colors=['firebrick','navy'],startangle=90,shadow=True,labels=['Fake','True'],autopct='%1.1f%%')
```

```python
# Train-Validation-Test set split into 70:15:15 ratio
# Train-Temp split
train_text, temp_text, train_labels, temp_labels = train_test_split(data['title'], data['label'],
                                                                    random_state=2018,
                                                                    test_size=0.25,
                                                                    stratify=data['Target'])
# Validation-Test split
val_text, test_text, val_labels, test_labels = train_test_split(temp_text, temp_labels,
                                                                random_state=2018,
                                                                test_size=0.5,
                                                                stratify=temp_labels)
```

```python
# Reshape input tensors to 2D
train_seq = tf.squeeze(train_seq)  # Remove extra dimensions
train_mask = tf.squeeze(train_mask)
val_seq = tf.squeeze(val_seq)
val_mask = tf.squeeze(val_mask)

# Recreate datasets with correct shapes
train_data = tf.data.Dataset.from_tensor_slices((train_seq, train_mask, train_y))
train_data = train_data.shuffle(buffer_size=10000).batch(batch_size).prefetch(tf.data.AUTOTUNE)

val_data = tf.data.Dataset.from_tensor_slices((val_seq, val_mask, val_y))
val_data = val_data.batch(batch_size).prefetch(tf.data.AUTOTUNE)

# Alternatively, if reshaping doesn't work, you can use tf.ensure_shape()
class BERT_Arch(tf.keras.Model):
    def __init__(self, bert, num_classes=2):
        super(BERT_Arch, self).__init__()
        self.bert = bert
        self.dropout = layers.Dropout(0.1)
        self.relu = layers.ReLU()
        self.fc1 = layers.Dense(512, activation=None)
        self.fc2 = layers.Dense(num_classes)
    
    def call(self, inputs, attention_mask=None, training=False):  
        # Ensure inputs are 2D
        inputs = tf.ensure_shape(inputs, [None, None])
        if attention_mask is not None:
            attention_mask = tf.ensure_shape(attention_mask, [None, None])
        
        # BERT output
        bert_output = self.bert(
            input_ids=inputs, 
            attention_mask=attention_mask, 
            training=training
        )
        cls_hs = bert_output.pooler_output  # Extract the pooled output
        
        # Pass through the fully connected layers
        x = self.fc1(cls_hs)
        x = self.relu(x)
        x = self.dropout(x, training=training)
        x = self.fc2(x)  # Output layer
        return x

# Modify train and evaluate functions to handle potential shape issues
def train(model, train_dataset, optimizer, loss_fn, batch_size):
    total_loss = 0
    total_accuracy = 0
    steps = 0
    
    for batch in train_dataset:
        sent_id, mask, labels = batch
        
        # Ensure correct shapes
        sent_id = tf.ensure_shape(sent_id, [None, None])
        mask = tf.ensure_shape(mask, [None, None])
        
        with tf.GradientTape() as tape:
            # Forward pass: Pass the sent_id and mask to the model
            preds = model(sent_id, attention_mask=mask, training=True)
            
            # Calculate the loss
            loss = loss_fn(labels, preds)
        
        # Calculate gradients and update the model
        gradients = tape.gradient(loss, model.trainable_variables)
        optimizer.apply_gradients(zip(gradients, model.trainable_variables))
        
        # Calculate accuracy
        accuracy = tf.keras.metrics.sparse_categorical_accuracy(labels, preds)
        
        total_loss += loss.numpy()
        total_accuracy += tf.reduce_mean(accuracy).numpy()
        steps += 1
        
    avg_loss = total_loss / steps
    avg_accuracy = total_accuracy / steps
    return avg_loss, avg_accuracy

def evaluate(model, val_dataset, loss_fn, batch_size=32):
    total_loss = 0
    total_accuracy = 0
    total_batches = 0
    
    for batch in val_dataset:
        sent_id, mask, labels = batch
        
        # Ensure correct shapes
        sent_id = tf.ensure_shape(sent_id, [None, None])
        mask = tf.ensure_shape(mask, [None, None])
        
        # Get predictions from the model
        preds = model(sent_id, attention_mask=mask, training=False)
        
        # Calculate the loss
        loss = loss_fn(labels, preds)
        total_loss += loss
        
        # For accuracy: compare predicted labels with true labels
        pred_labels = tf.argmax(preds, axis=1)
        accuracy = tf.reduce_mean(tf.cast(tf.equal(pred_labels, labels), tf.float32))
        total_accuracy += accuracy
        total_batches += 1
    
    # Calculate average loss and accuracy
    avg_loss = total_loss / total_batches
    avg_accuracy = total_accuracy / total_batches
    return avg_loss, avg_accuracy
```
